```python


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


np.random.seed(42)
num_vehicles = 3000
points_per_vehicle = 50

all_trajectories = []

for v_id in range(num_vehicles):
    x = np.cumsum(np.random.randn(points_per_vehicle))
    y = np.cumsum(np.random.randn(points_per_vehicle))
    speed = np.abs(np.random.randn(points_per_vehicle) * 5 + 30)
    acceleration = np.abs(np.random.randn(points_per_vehicle) * 0.5)
    turning_radius = np.abs(np.random.randn(points_per_vehicle) * 10 + 50)

    vehicle_df = pd.DataFrame({
        'vehicle_id': v_id,
        'x': x,
        'y': y,
        'speed': speed,
        'acceleration': acceleration,
        'turning_radius': turning_radius
    })

    all_trajectories.append(vehicle_df)

df = pd.concat(all_trajectories).reset_index(drop=True)
print(f" Simulated {num_vehicles} vehicle trajectories with {len(df)} total points.")


features = df[['speed', 'acceleration', 'turning_radius']]
kmeans = KMeans(n_clusters=4, random_state=42)
df['behavior_cluster'] = kmeans.fit_predict(features)

print(f" Clustering complete: {df['behavior_cluster'].nunique()} behavior clusters found.")


sample_size = 300  # adjust as needed for speed
sampled_ids = np.random.choice(df['vehicle_id'].unique(), sample_size, replace=False)
sampled_df = df[df['vehicle_id'].isin(sampled_ids)]


plt.figure(figsize=(10, 8))
for cluster in sorted(sampled_df['behavior_cluster'].unique()):
    subset = sampled_df[sampled_df['behavior_cluster'] == cluster]
    plt.plot(subset['x'], subset['y'], label=f'Cluster {cluster}', linewidth=1)

plt.title('AV Vehicle Trajectories Clustered by Behavior')
plt.xlabel('X position (meters)')
plt.ylabel('Y position (meters)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

print(" Project complete: Trajectory simulation, clustering, and visualization finished.")
```

```python
import matplotlib.cm as cm

plt.figure(figsize=(12, 8))


colors = cm.get_cmap('tab10', df['behavior_cluster'].nunique())

for cluster in sorted(sampled_df['behavior_cluster'].unique()):
    subset = sampled_df[sampled_df['behavior_cluster'] == cluster]
    plt.plot(
        subset['x'], subset['y'],
        label=f'Behavior {cluster}',
        linewidth=1.5,
        alpha=0.8,
        color=colors(cluster)
    )


plt.title('Autonomous Vehicle Trajectories by Behavior Cluster', fontsize=16)
plt.xlabel('X Position (meters)', fontsize=12)
plt.ylabel('Y Position (meters)', fontsize=12)
plt.legend(title='Cluster', fontsize=10)
plt.grid(True, linestyle='--', linewidth=0.5)
plt.tight_layout()


plt.savefig("av_behavior_clusters.png", dpi=300)

plt.show()
```

```python
import pandas as pd
import numpy as np

# Example trajectory dataset (replace this with your extracted data)
data = {
    'vehicle_id': np.repeat([1,2], 10),
    'time': list(range(10)) * 2,
    'x': np.random.randn(20).cumsum(),
    'y': np.random.randn(20).cumsum()
}
df = pd.DataFrame(data)

# Sort + detect missing time steps (fill with linear interpolation)
df = df.sort_values(by=['vehicle_id', 'time'])
df = df.groupby('vehicle_id').apply(lambda group: group.interpolate(method='linear')).reset_index(drop=True)
```

```python
def compute_kinematics(group):
    group = group.sort_values(by='time')
    group['dx'] = group['x'].diff()
    group['dy'] = group['y'].diff()
    group['dt'] = group['time'].diff().replace(0, np.nan)

    group['speed'] = np.sqrt(group['dx']**2 + group['dy']**2) / group['dt']
    group['acceleration'] = group['speed'].diff() / group['dt']


    group['curvature'] = np.abs(group['dy'] * group['dx'].shift(-1) - group['dx'] * group['dy'].shift(-1)) / \
                         ((group['dx']**2 + group['dy']**2)**1.5)
    group['turning_radius'] = 1 / (group['curvature'] + 1e-6)

    return group

df = df.groupby('vehicle_id').apply(compute_kinematics)
df = df.replace([np.inf, -np.inf], np.nan).fillna(0)
```

```python
from sklearn.cluster import KMeans


features = df[['speed', 'acceleration', 'turning_radius']].fillna(0)
model = KMeans(n_clusters=4, random_state=42)
df['behavior_cluster'] = model.fit_predict(features)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
for cluster in sorted(df['behavior_cluster'].unique()):
    subset = df[df['behavior_cluster'] == cluster]
    plt.plot(subset['x'], subset['y'], label=f'Cluster {cluster}')

plt.title('AV Vehicle Trajectories by Behavior Cluster')
plt.xlabel('X position (m)')
plt.ylabel('Y position (m)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df = df.reset_index(drop=True)  # 🪄 This solves the ambiguity

from sklearn.linear_model import LinearRegression

def predict_next_position(vehicle_group):
    model = LinearRegression()
    # Predict x based on time
    model.fit(vehicle_group['time'].values.reshape(-1,1), vehicle_group['x'].values)
    next_time = vehicle_group['time'].max() + 1
    next_x = model.predict([[next_time]])[0]
    # Predict y
    model.fit(vehicle_group['time'].values.reshape(-1,1), vehicle_group['y'].values)
    next_y = model.predict([[next_time]])[0]

    return pd.Series({'next_x': next_x, 'next_y': next_y})

predictions = df.groupby('vehicle_id').apply(predict_next_position)
print(predictions)
```
